```python
def NYdate(jula_year):
    ny = jula_year*0.25875-int(jula_year/4 +0.5)+int(jula_year/100 +0.38)-int(jula_year/400 +0.595)-5.53375
    jultin =  ((jula_year*292207)+373)%800
    hour,frac = divmod(jultin*24,800)
    minute,frac = divmod(frac*60,800)
    second,frac = divmod(frac*60,800)
    christ_year = jula_year+638
    dtny = datetime(christ_year, 4, int(ny), hour , minute, second)
    return dtny;
```

```python
def AS(dt):
    anto = [120,96,72,120,144,168,168,144,120,72,96,120]

    #dt = datetime(1998,9,18,17,8)
    #dt = datetime(1984,1,27,9,49)
    #dt = datetime(2009,4,16,5,36)
    offset = timedelta(minutes=18)
    dt= dt-offset
    dt6_00 = datetime(dt.year,dt.month,dt.day,6,0)
    
    if (dt>=dt6_00):
        dt_ref = dt6_00
    elif (dt<dt6_00):
        diff = timedelta(days=1)
        dt_ref = dt6_00 - diff

    #hj1 = hora(dt_ref)
    hj1 = hora(dt_ref+offset)
    sun6_0 = hj1.sun

    z = sun6_0//1800 
    m = (sun6_0%1800)

    FirstpartialRemain = (1800-m)*anto[z]/1800
    timediff = dt - dt_ref

    totalMinute,frac = divmod(timediff.total_seconds(), 60)
    if frac != 0: raise Exception('seconds not equal to 0')   


    if totalMinute < FirstpartialRemain:
        zodiac = z
        lip = m + (totalMinute*1800)/anto[zodiac]
    else:    
        TotalTimeRemain = totalMinute-FirstpartialRemain
        n = z+1
        while (TotalTimeRemain >= anto[n%12]):
            TotalTimeRemain -= anto[n%12]
            n+=1
        zodiac = n%12
        lip = (TotalTimeRemain*1800)/anto[zodiac]

    lip = int(round(lip))
    total_lip = zodiac*1800+lip
    return total_lip

#     print('dt =',dt)
#     print('dt6_00 =',dt6_00)
#     print('dt_ref =',dt_ref)
#     print('sun6_0 =',sun6_0)
#     print('z =',z)
#     print('m =',m)
#     print('FirstpartialRemain =',FirstpartialRemain)
#     print('timediff =',timediff)
#     print('totalMinute =',totalMinute)
#     print('frac =',frac)
#     # print('a =',a)
#     # print('b =',b)
#     # print('c =',c)
#     # print('d =',d)
#     # print('e =',e)
#     print('TotalTimeRemain =',TotalTimeRemain)
#     print('n =',n)
#     print('lip =',lip)
#     print('total_lip =',total_lip)
#     print('zodiac =',zodiac)
#     print('degree =',degree)
#     print('minute =',minute)



    
```
